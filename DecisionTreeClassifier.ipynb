```python
import pandas as pd
df = pd.read_csv('2August.csv')
df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X = df.drop(columns=['Disease'])
y = df['Disease']
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
prediction = model.predict([[1, 45, 110, 4.5, 0.8, 13.1, 13.2, 150.1, 11.1, 84.5, 18.4, 103.5, 43.1, 110.8, 43.1, 0.34, 0.58, 1.12]])
prediction
```
