```python
import pandas as pd
df = pd.read_csv('1August.csv')
df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X = df.drop(columns=['Disease'])
y = df['Disease']
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
prediction = model.predict([[1, 65, 136, 6.9, 0.5, 12, 13.4, 116.4, 3.1, 62.9, 5.6, 98.0, 23.4, 103.5, 4.5, 0.34, 0.81, 1.201]])
prediction
```
